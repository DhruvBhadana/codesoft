```python
from tkinter import *

# Main window
window = Tk()
window.title("Simple Calculator")
window.geometry('400x300')

def calculate():
    try:
        num1 = float(entry_num1.get())
        num2 = float(entry_num2.get())
    except ValueError:
        result_var.set("Invalid input!")
        return
    
    operation = operation_var.get()
    
    if operation == "+":
        result = num1 + num2
    elif operation == "-":
        result = num1 - num2
    elif operation == "*":
        result = num1 * num2
    elif operation == "/":
        if num2 == 0:
            result_var.set("Error: Division by zero")
            return
        result = num1 / num2
    else:
        result_var.set("Select an operation")
        return
    
    result_var.set(f"Result: {result}")

# First number input
label_num1 = Label(window, text="Enter first number:")
label_num1.pack(pady=5)
entry_num1 = Entry(window)
entry_num1.pack(pady=5)

# Second number input
label_num2 = Label(window, text="Enter second number:")
label_num2.pack(pady=5)
entry_num2 = Entry(window)
entry_num2.pack(pady=5)

# Operation selection
label_operation = Label(window, text="Choose an operation:")
label_operation.pack(pady=5)
operation_var = StringVar()
operation_var.set("+")

frame_operations = Frame(window)
frame_operations.pack(pady=5)

radio_add = Radiobutton(frame_operations, text="+", variable=operation_var, value="+")
radio_add.pack(side=LEFT, padx=5)

radio_subtract = Radiobutton(frame_operations, text="-", variable=operation_var, value="-")
radio_subtract.pack(side=LEFT, padx=5)

radio_multiply = Radiobutton(frame_operations, text="*", variable=operation_var, value="*")
radio_multiply.pack(side=LEFT, padx=5)

radio_divide = Radiobutton(frame_operations, text="/", variable=operation_var, value="/")
radio_divide.pack(side=LEFT, padx=5)

# Calculate button
button_calculate = Button(window, text="Calculate", command=calculate)
button_calculate.pack(pady=10)

# Result display
result_var = StringVar()
result_var.set("Result: ")
label_result = Label(window, textvariable=result_var)
label_result.pack(pady=5)

window.mainloop()
```
